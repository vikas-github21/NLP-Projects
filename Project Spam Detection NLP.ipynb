# SPAM Detection :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
sms = pd.read_csv('spam.csv' , encoding ='latin-1')
sms.head()
```

```python
sms.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
sms.rename(columns={'v1':'label','v2':'message'},inplace=True)
```

```python
sms.head()
```

```python
print('Shape = >',sms.shape)
```

```python
print('ham and spam counts', '\n', sms.label.value_counts())
```

```python
print ('spam ratio =', round(len(sms[sms['label']=='spam']) / len(sms.label),2)*100,'%')
print ('ham ratio =', round(len(sms[sms['label']=='ham']) / len(sms.label),2)*100,'%')
```

```python
sms['length'] = sms.message.str.len()
sms.head(5)
```

```python
sms['label'].replace({'ham':0,'spam':1},inplace=True)
```

```python
sms.head(5)
```

```python
sms['message'] = sms['message'].str.lower()
```

```python
sms.head()
```

## Regular Expression

```python
# Replace email address with 'email'
sms['message'] = sms['message'].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$','emailaddress')

# Reolace URL present in message with webaddress
sms['message'] = sms['message'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$','webaddress')

# Replace pound symbol with dollers
sms['message'] = sms['message'].str.replace(r'Â£|\$', 'dollers')

# Replace phonenumbers with phonenumber
sms['message'] = sms['message'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$','phonenumber')

# Replace any number present in message with numbr
sms['message'] = sms['message'].str.replace(r'\d+(\.\d+)?', 'numbr')
```

```python
#Remove Punctuation
sms['message'] = sms['message'].str.replace(r'[^\w\d\s]', ' ')

#replace white space between terms with a single space
sms['message'] = sms['message'].str.replace(r'\s+', ' ')

# remove loading and trailing whitespace
sms['message'] = sms['message'].str.replace(r'^\s+|\s+?$', '')
```

```python
sms.head()
```

```python
# Remove Stopwords
import string
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english') +['u', 'ur','4','2','im', 'dont', 'doin','ure'])

sms['message'] = sms['message'].apply(lambda x: ' '.join(term for term in x.split() if term not in stop_words))
```

### Creating New Column Clean Length After Removing Stopwords

```python
sms['clean_length'] = sms.message.str.len()
sms.head()
```

```python
print('Original Length', sms.length.sum())
print('Clean Length', sms.clean_length.sum())
```

```python
# Message distribution BEFORE Cleaning

f,ax = plt.subplots(1,2,figsize =(15,8))
sns.distplot(sms[sms['label']==1]['length'],bins=20,ax=ax[0],label='Spam messages distribution',color='r')

ax[0].set_xlabel('Spam sms length')
ax[0].legend

sns.distplot(sms[sms['label']==0]['length'],bins=20,ax=ax[1],label='ham messages distribution')

ax[1].set_xlabel('ham sms length')
ax[1].legend()

plt.show
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report

tf_vec = TfidfVectorizer()
naive =  MultinomialNB()

features = tf_vec.fit_transform(sms['message'])

X = features
y = sms['label']
```

```python
X_train,x_test,Y_train,y_test = train_test_split(X,y,random_state=42)
naive.fit(X_train,Y_train)
y_pred=naive.predict(x_test)

print('Final score = > ', accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
conf_mat = confusion_matrix(y_test,y_pred)
ax=plt.subplot()

sns.heatmap(conf_mat,annot=True,ax=ax,linewidths=5,linecolor='r',center=0)
ax.set_xlabel('Predicted Lables');ax.set_ylabel('True Lables')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['ham','spam'])
ax.yaxis.set_ticklabels(['ham','spam'])

plt.show
```
